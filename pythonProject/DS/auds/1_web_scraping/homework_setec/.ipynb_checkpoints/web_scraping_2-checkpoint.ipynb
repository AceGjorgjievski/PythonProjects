```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url = "https://setec.mk"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
categories = soup.select(".with-sub-menu")
len(categories)/2
```

```python
all_categories = []

for category in categories:
    all_categories.append(category)
```

```python
category_links = {}
```

```python
for i in range(int(len(all_categories)/2), len(all_categories)): 
    """
        len(all_categories)/2 because the dropdown modal has a left and 
        a right side of the same categories twice, that's why
    """
    
    # print(all_categories[i].select_one(".clearfix").text)
    category_name = all_categories[i].select_one(".clearfix").text
    col_sm_elements = all_categories[i].select("div.row div[class^='col-sm'] a")

    if category_name not in category_links:
        category_links[category_name] = []
    
    for col in col_sm_elements:
        href = col["href"]

        if href.startswith("index.php"):
            href = "https://setec.mk/" + href
        
        category_links[category_name].append(href)
        # print(col["href"])
```

### Scraping category page

```python
parsed_products = []
```

```python
def scrape_products_per_link(category, link):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, "html.parser")
    products = soup.select(".product")
    for product in products:
        name = product.select_one(".name").text.strip()
        code = product.select_one(".shifra").text.strip()
        regular_price = product.select_one(".price-old-new")
        if regular_price:
            regular_price = product.select_one(".price-old-new").text.strip()

        
        happy_price = product.select_one(".price-new-new")

        if happy_price:
            happy_price = product.select_one(".price-new-new").text.strip()
        
        discount = product.select_one(".sale")

        if discount:
            discount = product.select_one(".sale").text.strip()
        
        link = product.select_one(".image a").get("href")
        image_link = product.find("img").get("data-echo")

        product_dict = {
            "Product Name" : name,
            "Product Code" : code,
            "Regular Price" : regular_price,
            "Happy Price" : happy_price,
            "Discount": discount,
            "Page Link": link,
            "Image Link": image_link,
            "Category": category
        }

        """
                I print out every product when it adds up in the list, 
                in order to see how many left... because it takes around 2-3 mins in order
                to fulfill the list
        """
        
        
        print(product_dict) 
        parsed_products.append(product_dict)
```

```python
#scrape_products_per_link("PC", "http://midea.mk/?page_id=3425") #example if it works
```

```python
for category, links in category_links.items():
    if category == 'Професионална Климатизација': 
        """
        I skipped this category cuz it navigates you to another site, and there
        another logic for scraping.... if you want to scrape there as well, check these links:
            http://midea.mk/?page_id=3425
            http://midea.mk/?page_id=8105
            http://midea.mk/?page_id=7082
            http://midea.mk/?page_id=3409
            http://midea.mk/?page_id=3431
            http://midea.mk/?page_id=5050
            http://midea.mk/?page_id=3409
            http://midea.mk/
            """
        continue
    for link in links:
        scrape_products_per_link(category, link)
```
