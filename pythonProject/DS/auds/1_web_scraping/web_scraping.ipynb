```python
all_products = soup.select(".product")
```

```python
print(all_products[0])
```

```python
first_product = all_products[0]
```

```python
name = first_product.select_one(".name").text.strip()
code = first_product.select_one(".shifra").text.strip()
regular_price = first_product.select_one(".price-old-new").text.strip()
happy_price = first_product.select_one(".price-new-new").text.strip()
discount = first_product.select_one(".sale").text
link = first_product.select_one(".image a").get("href")
image_link = first_product.find("img").get("data-echo")
```

```python
name, code, link, image_link
```

```python
parsed_products = []

for product in all_products:
    name = first_product.select_one(".name").text.strip()
    code = first_product.select_one(".shifra").text.strip()
    regular_price = first_product.select_one(".price-old-new").text.strip()
    happy_price = first_product.select_one(".price-new-new").text.strip()
    discount = first_product.select_one(".sale").text
    link = first_product.select_one(".image a").get("href")
    image_link = first_product.find("img").get("data-echo")

    product_dict = {
        "Product Name" : name,
        "Product Code" : code,
        "Regular Price" : regular_price,
        "Happy Price" : happy_price,
        "Discount": discount,
        "Page Link": link,
        "Image Link": image_link
    }

    parsed_products.append(product_dict)

```

```python
parsed_products
```

```python
df = pd.DataFrame(parsed_products)
```
